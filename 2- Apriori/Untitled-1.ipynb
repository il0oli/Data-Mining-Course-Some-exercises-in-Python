```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# قراءة البيانات من ملف CSV
df = pd.read_csv('dataset/breadBasket.csv')
```

```python
# عرض أول 5 صفوف من البيانات للتحقق
print(df.head())
```

```python
transactions_str = df.groupby(['Transaction', 'Item'])['Item'].count().reset_index(name ='Count')
transactions_str
```

```python
# إنشاء مصفوفة mxn حيث m=معاملة و n=عناصر وكل صف يمثل ما إذا كان العنصر موجودًا في المعاملة أم لا

my_basket = transactions_str.pivot_table(index='Transaction', columns='Item', values='Count', aggfunc='sum').fillna(0)
```

```python
my_basket.head()
```

```python
# إنشاء دالة ترجع 0 أو 1
# 0 يعني أن العنصر لم يكن في هذه المعاملة، 1 يعني أن العنصر موجود في هذه المعاملة

def encode(x):
    if x<=0:
        return 0
    if x>=1:
        return 1

# applying the function to the dataset

my_basket_sets = my_basket.applymap(encode)
my_basket_sets.head()
```

```python
# باستخدام "خوارزمية apriori" مع min_support=0.01 (1% من 9465)
# وهذا يعني أن العنصر يجب أن يكون موجودًا في 94 معاملة على الأقل من أصل 9465 معاملة فقط عندما نأخذ هذا العنصر في الاعتبار
# مجموعة العناصر المتكررة

frequent_items = apriori(my_basket_sets, min_support = 0.01,use_colnames = True)
frequent_items
```

```python
# الآن يتم إنشاء القواعد من مجموعة العناصر المتكررة التي تم إنشاؤها أعلاه

rules = association_rules(frequent_items, num_itemsets = 60 , metric = "lift", min_threshold = 1)
rules.sort_values('confidence', ascending = False, inplace = True)
rules
```

```python
# arranging the data from highest to lowest with respect to 'confidence'

rules.sort_values('confidence', ascending=False)
```

```python
# عرض القواعد
print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
