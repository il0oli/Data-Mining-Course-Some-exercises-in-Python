```python
# تحليل القيم المفقودة
missing_columns = data.columns[data.isnull().any()]
print("Columns with missing values:")
display(data[missing_columns].isnull().sum())
```

```python
# معالجة القيم المفقودة
for column in missing_columns:
    if data[column].dtype == 'object':
        # إذا كان العمود نصيًا
        most_common_value = data[column].mode()[0] if not data[column].mode().empty else "Unknown"
        data[column].fillna(most_common_value, inplace=True)
        print(f"Filled missing values in {column} with most common value: {most_common_value}")
    elif data[column].dtype in ['float64', 'int64']:
        # إذا كان العمود رقميًا
        mean_value = data[column].mean()
        data[column].fillna(mean_value, inplace=True)
        print(f"Filled missing values in {column} with mean value: {mean_value}")
```

```python
# التأكد من نظافة البيانات
print("\nMissing values after cleaning:")
display(data.isnull().sum())
```
